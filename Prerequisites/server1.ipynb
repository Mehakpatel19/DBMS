```python
import socket
#server code
# Server configuration
HOST = "127.0.0.1"  # Localhost
PORT = 5000

object1 = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
object1.bind((HOST, PORT))
print(f"UDP Server listening on {HOST}:{PORT}")

while True:
    data, client_address = object1.recvfrom(1024)
    message = data.decode()
    print(f"Received from {client_address}: {message}")

    response = f"Hello Client"
    object1.sendto(response.encode(), client_address)


object1.close()
```
