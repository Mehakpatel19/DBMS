```python
import socket
import time
#client 1 ke liye
# Server configuration
SERVER_HOST = "127.0.0.1"
SERVER_PORT = 5000

HOST = "127.0.0.1"  # Localhost
PORT = 6000

object2 = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
object2.bind((HOST, PORT))

message = "Hello i am client 1"

while True:
    object2.sendto(message.encode(), (SERVER_HOST, SERVER_PORT))
    response, _ = object2.recvfrom(1024)
    print("Server says : ", response.decode())
    time.sleep(1)

object2.close()
```
